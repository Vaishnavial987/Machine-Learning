```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import  Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import  RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from  sklearn.metrics import r2_score
import seaborn as sns
from sklearn.svm import  SVR
```

```python
df=pd.read_csv('train.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.drop(columns=['Open Date'])
```

```python
df['City'].unique()
```

```python
df['City Group'].unique()
```

```python
df['Type'].unique()
```

```python
x=df.drop('revenue',axis=1)
y=df.revenue
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
num_col=xtrain.select_dtypes(include='number').columns
obj_col=xtrain.select_dtypes(exclude='number').columns
```

```python
preprocessing=ColumnTransformer(
    transformers=[
        ('onehotencoder',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),obj_col.drop('City Group')),
        ('ordinalencoder',OrdinalEncoder(),['City Group'])
    ]
)
```

```python
decision_pipeline=Pipeline(
    steps=[
        ('preprocessing',preprocessing),
        ('model',DecisionTreeRegressor())
    ]
)
```

```python
decision_pipeline.fit(xtrain,ytrain)
```

```python
print("Train_score:",decision_pipeline.score(xtrain,ytrain))
print("Test_score:",decision_pipeline.score(xtest,ytest))
```

```python
Random_forest_pipeline=Pipeline(
    steps=[
('preprocessing',preprocessing),
('model',RandomForestRegressor())
    ]
)
```

```python
Random_forest_pipeline.fit(xtrain,ytrain)
```

```python
print("train_score:",Random_forest_pipeline.score(xtrain,ytrain))
print("test_score:",Random_forest_pipeline.score(xtest,ytest))
```

```python
bagging_pipeline=Pipeline(
    steps=[
('preprocessing',preprocessing),
('model',BaggingRegressor(estimator=SVR()))
    ]
)
```

```python
bagging_pipeline.fit(xtrain,ytrain)
```

```python
print("train_score:",bagging_pipeline.score(xtrain,ytrain))
print("test_score:",bagging_pipeline.score(xtest,ytest))
```
