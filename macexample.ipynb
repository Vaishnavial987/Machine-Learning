```python
import pandas as pd
import numpy as np
```

```python
s={'Temperature':[35,33,30,28,27],'Ie_cream_sales':[5000,4000,3000,2000,500]}
print(s)
```

```python
df=pd.DataFrame(s)
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=df['Temperature']  
y=df['Ie_cream_sales'] 
```

```python
mx=np.mean(x)     #mean of x
mx
```

```python
my=np.mean(y)    #mean of y
my
```

```python
vx=np.var(x)    #varience of x
vx
```

```python
vy=np.var(y)    #varience of y
vy
```

```python
sx=np.std(x)    #standard deviation of x
```

```python
sy=np.std(y)    #standard deviation of y
```

```python
cvx=sx/mx       #covarience of x
print(cvx*100)
```

```python
cvy=sy/my       #covarience of y
print(cvy*100)
```

```python
cv=np.cov(x,y)    #covarience of x,y
cv
```

```python
m=cv/vx        #slope
m
```

```python
c=my-m*mx   #intercept
c
```

```python
yp=m*mx+c   #predicted value
yp
```
